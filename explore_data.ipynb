```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
```

# Load data

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

# Examine train data

```python
_, axes = plt.subplots(4,7)
for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[i], cmap='gray_r')
    ax.set_title(y_train[i])
    ax.set_axis_off()
```

```python
plt.boxplot(x_train.flatten())
```

```python
plt.hist(x_train.flatten())
```

```python
plt.hist(y_train, bins=1)
```

# Examine test data

```python
_, axes = plt.subplots(4,7)
for i, ax in enumerate(axes.flat):
    ax.imshow(x_test[i], cmap='gray_r')
    ax.set_title(y_test[i])
    ax.set_axis_off()
```

```python
plt.boxplot(x_test.flatten())
```

```python
plt.hist(x_test.flatten())
```

```python
plt.hist(y_test, bins=1)
```

# Preprocess data

```python
x_train[x_train < 127] = 0
x_train[x_train >= 127] = 1

x_test[x_test < 127] = 0
x_test[x_test >= 127] = 1
```
